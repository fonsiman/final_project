```python
import os
import cv2
import numpy as np
from keras.utils import to_categorical
from imutils import rotate_bound
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
myFolder = "images"
imagepaths = set() 

for root, dirs, files in os.walk(myFolder):
    for fileName in files:
        imagepaths.add( os.path.join( root[len(myFolder):], fileName ))
```

```python
print(len(imagepaths))
print(imagepaths)
```

```python
X = [] # Image data
y = [] # Labels# Loops through imagepaths to load images and labels into arrays
i = 0
for path in imagepaths:
    img = cv2.imread(myFolder + path) # Reads image and returns np.array
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # Converts into the corret colorspace (GRAY)
    img = cv2.resize(img, (100,100)) # Reduce image size so training can be faster
    X.append(img)

    category = path.split("/")[1]
    label = int(category[0]) # We need to convert 10_down to 00_down, or else it crashes
    y.append(label)
    print(label)


print(len(X))
X = np.array(X, dtype="uint8")
print(len(imagepaths)*4)
X = X.reshape(len(X), 100, 100, 1) # Needed to reshape so CNN knows it's different images
y = np.array(y)
print("Images loaded: ", len(X))
print("Labels loaded: ", len(y))
```

```python
import pickle

pickle_out = open('pickles/X2.pickle','wb')
pickle.dump(X, pickle_out)
pickle_out.close()

pickle_out = open('pickles/y2.pickle','wb')
pickle.dump(y, pickle_out)
pickle_out.close()
```

```python
X_train, X_aux, y_train, y_aux = train_test_split(X, y, test_size=.3, random_state=42, stratify=y)

X_test, X_val, y_test, y_val = train_test_split(X_aux, y_aux, test_size=.5, random_state=42, stratify=y_aux)
```

```python
from keras.models import Sequential
from keras.layers.convolutional import Conv2D, MaxPooling2D
from keras.layers import Dense, Flatten, Dropout
from keras.optimizers import Adadelta


model = Sequential()
model.add(Conv2D(filters=32, kernel_size=(3, 3), input_shape=(100, 100, 1), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.2))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(units=4, activation='softmax'))
model.compile(
    loss='categorical_crossentropy',
    optimizer=Adadelta(),  
    metrics=['accuracy'])
```

```python
model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, to_categorical(y_train), epochs=50, batch_size=64, verbose=2, validation_data=(X_val, to_categorical(y_val)))
```
